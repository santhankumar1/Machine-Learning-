```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
iris
```

```python
print(iris['DESCR'])
```

```python
print(iris['target'])
```

```python
x=pd.DataFrame(iris['data'],columns=['sepal length in cm',
     'sepal width in cm',
    'petal length in cm',
     'petal width in cm'])
```

```python
x
```

```python
y=iris['target']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=23)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
tree1=DecisionTreeClassifier(max_depth=2)
tree1.fit(x_train,y_train)
y_predict=tree1.predict(x_test)

```

```python
#Visualize decision tree
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(tree1,filled=True)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
score1=accuracy_score(y_predict,y_test)
print(score1)
score2=classification_report(y_predict,y_test)
print(score2)
```

```python
score4=confusion_matrix(y_predict,y_test)
print(score4)
```

# #Hyper Parameter tunning

```python
paramas_grid={
    'max_features' :  ["sqrt", "log2","auto"],
'max_depth' : [1,2,4,5,7,10,11],
'splitter' : ["best", "random"],
'criterion' : ["gini", "entropy", "log_loss"]
    
}
```

```python
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(DecisionTreeClassifier(),param_grid=paramas_grid,
    scoring='accuracy',
    n_jobs=-1,
    refit=True,
    cv=3,
    verbose=3)
grid.fit(x_train,y_train)
y_pre= grid.predict(x_test)
print(y_pre)
```

```python
grid.best_score_
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
score=confusion_matrix(y_test,y_pre)
score1=classification_report(y_test,y_pre)
score2=accuracy_score(y_test,y_pre)
print(score)
print(score1)
print(score2)
```
