#Train Test SPlit and Model Training

```python
from sklearn.model_selection import train_test_split
x=df1.drop(columns=['ProdTaken'],axis=1)# Independent Features
y=df1['ProdTaken']# Dependent Feature
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=11)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
# Create columns transform with 3dtypes  of transformers
cat_features=x.select_dtypes(include='object').columns
num_features=x.select_dtypes(exclude='object').columns

from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.compose import ColumnTransformer

preprocessing=ColumnTransformer(transformers=([
    ('OneHotEncoder',OneHotEncoder(drop='first'),cat_features),
    ('Standardscaler',StandardScaler(),num_features)
]))
```

```python
x_train=preprocessing.fit_transform(x_train)
```

```python
x_test=preprocessing.transform(x_test)
```

```python
pd.DataFrame(x_train)
```

# RandomForestClassifier Training

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,recall_score,precision_score,f1_score,roc_auc_score,roc_curve
```

```python
model={
    "Randomforest":RandomForestClassifier(),
    "DecisionTree":DecisionTreeClassifier(),
    "Logisticregression":LogisticRegression(),
    "AdaBoostClassifier":AdaBoostClassifier()
}
for i in range(len(list(model))):
    models=list(model.values())[i]
    models.fit(x_train,y_train)#Train Model
    
    #Model Predictions
    y_train_pred=models.predict(x_train)
    y_test_pred=models.predict(x_test)
    
    #Training set Performance
    model_train_acc=accuracy_score(y_train,y_train_pred)
    model_train_f1=f1_score(y_train,y_train_pred,average='weighted')
    model_train_pre=precision_score(y_train,y_train_pred)
    model_train_recall=recall_score(y_train,y_train_pred)
    model_train_roc=roc_auc_score(y_train,y_train_pred)
    
    #test set performance
    model_test_acc=accuracy_score(y_test,y_test_pred)
    model_test_f1=f1_score(y_test,y_test_pred,average="weighted")
    model_test_pre=precision_score(y_test,y_test_pred)
    model_test_recall=recall_score(y_test,y_test_pred)
    model_test_roc=roc_auc_score(y_test,y_test_pred)
    print('-'*45)
    print(list(model.keys())[i])
    print('-'*45)
    
    
    print('Model Performace for traning set')
    print('Accuracy score:{:.4f}'.format(model_train_acc))
    print('F1 Score:{:.4f}'.format(model_train_f1))
    print('Precision:{:.4f}'.format(model_train_pre))
    print('recall:{:.4f}'.format(model_train_recall))
    print('roc:{:.4f}'.format(model_train_roc))
    

    
    print('Accuracy_score:{:.4f}'.format(model_test_acc))
    print('F1 score:{:.4f}'.format( model_test_f1))
    print('Precision:{:.4f}'.format(model_test_pre))
    print('Recall:{:.4f}'.format(model_test_recall))
    print('roc:{:.4f}'.format(model_test_roc))
    
```

```python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.model_selection import RandomizedSearchCV

# Define hyperparameter grids
rf_params = {
    'criterion': ["gini", "entropy", "log_loss"],
    'n_estimators': [100, 200, 400, 500, 700, 900, 1000],
    'max_depth': [2, 4, 8, 12, 16],
    'max_features': [1, 3, 5, 8, 10, 12],
    'max_leaf_nodes': [1, 4, 6, 8, 10],
    'min_samples_split': [1, 2, 4, 8, 10],
    'class_weight': ["balanced", "balanced_subsample"]
}

ada_params = {
    "n_estimators": [50, 60, 70, 80, 100],
    "algorithm": ['SAMME']
}

# List of models with their parameters
randomcv_models = [
    ("RF", RandomForestClassifier(), rf_params),
    ("ADA", AdaBoostClassifier(), ada_params)
]

# Initialize dictionary to store the best parameters
model_param = {}

# Perform Randomized Search for each model
for name, model, params in randomcv_models:
    try:
        print(f'Running RandomizedSearchCV for {name}...')
        random = RandomizedSearchCV(
            estimator=model,
            param_distributions=params,
            n_iter=100,
            n_jobs=-1,
            cv=3
        )
        random.fit(x_train, y_train)
        model_param[name] = random.best_params_
    except Exception as e:
        print(f'Error occurred while fitting {name}: {e}')

# Print the best parameters for each model
for mode_name in model_param:
    print(f'_______________________Best Params for {mode_name} ------------------------------')
    print(model_param[mode_name])

if 'RF' not in model_param:
    print("Random Forest model did not complete successfully.")
```
