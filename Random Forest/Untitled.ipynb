```python
# Independent and Dependent Features
x=df.iloc[:,:-1]  #Independent Features
y=df['target']#Dependent Features
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=11)
```

```python
Numerical_Features=x.select_dtypes(exclude='object').columns
print(Numerical_Features)
```

```python
#Data Transformer 
```

```python

from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer

```

```python
columnstransformer=ColumnTransformer(
    transformers=[('standard',StandardScaler(),Numerical_Features)],
    remainder='drop'
)

x_train=columnstransformer.fit_transform(x_train)
x_test=columnstransformer.transform(x_test)
```
