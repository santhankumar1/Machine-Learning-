#Logistic Regression for multiclass classification problem

```python
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=1000,n_features=10,n_classes=3,random_state=42,n_informative=3)
```

```python
from sklearn.model_selection import train_test_split
x_train1,x_test1,y_train1,y_test1= train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
regress=LogisticRegression(multi_class= 'ovr')
```

```python
regress.fit(x_train,y_train)
```

```python
y_pred1=regress.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
score1=accuracy_score(y_test,y_pred1)
print(score1)
confusion1=confusion_matrix(y_test,y_pred1)
print(confusion1)
print(classification_report(y_test,y_pred1))
```

```python
model=LogisticRegression()
```

```python
penalty=['l1', 'l2', 'elasticnet']
solver=['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
c_Values=[100,10,1.0,0.1,0.01]
```

```python
paramas=dict(penalty=penalty,C=c_Values,solver=solver)
```

```python
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold(n_splits=3)
```

```python
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(estimator=model,param_grid=paramas,scoring='accuracy',n_jobs=1,cv=cv)
```

```python
grid.fit(x_train1,y_train1)
```

```python
y_predict1=grid.predict(x_test)
```

```python
score1=accuracy_score(y_test,y_predict1)
print(score1)
confusion1=confusion_matrix(y_test,y_predict1)
print(confusion1)
print(classification_report(y_test,y_predict1))
```
