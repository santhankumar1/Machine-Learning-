```python


# Calculate the correlation matrix
corr = df1.corr()

# Iterating through the columns
for col in df1.columns:
    print("*" * 30)
    print(f"Evaluating correlations with {col}")

    # Ensure col is in the correlation matrix
    if col in corr.columns:
        try:
            # Drop self-correlation
            corr_index = corr[col].drop(index=col)
            print(f"Relevant correlations with {col}: {corr_index}")
            for other_col, value in corr_index.items():
                if abs(value) > 0.3:
                    print(f"{other_col}: {value}")
        except TypeError as e:
            print(f"Error accessing correlations for {col}: {e}")
```
